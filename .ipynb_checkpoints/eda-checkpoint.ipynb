### Usando dados do sitema de energia aberto para entender a serie temporal dos dados

- *Data* : A data está no formato yyyy-mm-dd.
- *Consumption* : Indica o consumo de eletricidade em GWh.
- *Solar* : Indica a produção de energia solar em GWh.
- *wind* + Solar : Representa a soma da produção de energia solar e eólica em GWh.

Questão levantada para entender os dados:

- Como o consumo e a produção de eletricidade variam ao longo do tempo na Alemanha?

#### Imports 


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load time series dataset
df_power = pd.read_csv("https://raw.githubusercontent.com/jenfly/opsd/master/opsd_germany_daily.csv")
df_power.columns
```

```python
df_power.tail(10)
```

#### Limpeza de Dados

```python
df_power.shape
```

```python
df_power.dtypes
```

```python
#converter o formado object para datetime
df_power['Date'] = pd.to_datetime(df_power['Date'])
```

```python
df_power.dtypes
```

```python
#definindo a coluna Date como Indice do dataframe, por que na analise de series temporais a coluna de Indice é sempre uma coluna datetime
df_power = df_power.set_index('Date')
df_power.tail(3)
```

```python
df_power.index
```

```python
# Adicionando colunas para Mes,ano e nome do dia da semana
df_power['Year'] = df_power.index.year
df_power['Month'] = df_power.index.month
df_power['Weekday Name'] = df_power.index.day_name()
```

```python
df_power.sample(5,random_state=0)
```

#### Indexação baseada em Tempo

```python
df_power.loc['2015-10-02']
```

```python
df_power.loc['2017-01-01':'2017-12-30']
```

### Visualização para series Temporais

```python
sns.set(rc={'figure.figsize':(16,4)})
plt.rcParams['figure.dpi'] = 150
```

Criando um grafico linear de serie temporal do consumo diario completo 

```python
df_power['Consumption'].plot(linewidth=0.4)
```

Usando os pontos para representar graficamente os dados de todas as colunas

```python
cols_to_plot = ['Consumption', 'Solar', 'Wind']
axes = df_power[cols_to_plot].plot(marker='.', alpha=0.5, linestyle='None',figsize=(14, 7), subplots=True)
for ax in axes:
    ax.set_ylabel('Daily Totals (GWh)')
```

- O resultado mostra que o consumo de eletricidade pode ser dividido em dois padrões distintos:

Um cluster com aproximadamente 1.400 GWh e acima
Outro cluster aproximadamente abaixo de 1.400 GWh
Além disso, a produção solar é maior no verão e menor no inverno. Ao longo dos anos, parece ter havido uma forte tendência crescente na produção de energia eólica.

```python
ax = df_power.loc['2016', 'Consumption'].plot()
ax.set_ylabel('Daily Consumption (GWh)');
```

```python
ax = df_power.loc['2016-12-23':'2016-12-30', 'Consumption'].plot(marker='o', linestyle='-')
ax.set_ylabel('Daily Consumption (GWh)');
```

- o consumo de electricidade foi mais baixo no dia de Natal, provavelmente porque as pessoas estavam ocupadas em festas. Depois do Natal, o consumo aumentou

```python
ax = df_power.loc['2016-12', 'Consumption'].plot(marker='o', linestyle='-')
ax.set_ylabel('Daily Consumption (GWh)');
```

### Sazonalidade

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 7), sharex=True)
palette = sns.color_palette("husl", 12) 

for name, ax in zip(['Consumption', 'Solar', 'Wind'], axes):
    sns.boxplot(data=df_power, x='Month', y=name, hue='Month', palette=palette, ax=ax, legend=False)
    ax.set_ylabel('GWh')
    ax.set_title(name)
    if ax != axes[-1]:
        ax.set_xlabel('')

plt.tight_layout()
plt.show()
```

#### Podemos ver que o consumo de electricidade é geralmente mais elevado no Inverno e mais baixo no Verão. A produção eólica é maior durante o verão. Além disso, existem muitos valores discrepantes associados ao consumo de eletricidade, à produção eólica e à produção solar.

```python
palette = sns.color_palette("husl", len(df_power['Weekday Name'].unique()))
sns.boxplot(data=df_power, x='Weekday Name', y='Consumption', hue='Weekday Name', palette=palette, dodge=False)


```

- o consumo de eletricidade é maior durante a semana do que nos finais de semana. Curiosamente, há mais valores discrepantes durante a semana

### Realizando uma reamostragem no conjunto de dados com base na serie temporal média semanal

```python
columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']

power_weekly_mean = df_power[columns].resample('W').mean()
power_weekly_mean.head(10)
     
```

* A primeira linha, rotulada como 2006-01-01, inclui a média de todos os dados. Podemos traçar as séries temporais diárias e semanais para comparar o conjunto de dados ao longo do período de seis meses.

```python
start, end = '2016-01', '2016-06'
```

```python
fig, ax = plt.subplots()

ax.plot(df_power.loc[start:end, 'Solar'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(power_weekly_mean.loc[start:end, 'Solar'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.set_ylabel('Solar Production in (GWh)')
ax.legend();
```

- a série temporal média semanal está aumentando ao longo do tempo e é muito mais suave do que a série temporal diária.
